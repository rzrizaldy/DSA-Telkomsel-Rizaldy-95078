### data is clean

```python
#1. Customer yang tidak melakukan service sama sekali


df_churn['Total Transaction'] = df_churn.apply(lambda row: row['total_day_minutes'] + row['total_day_calls'] + row['total_day_charge'] + row['total_eve_minutes'] + row['total_eve_calls'] + row['total_eve_charge'] + row['total_night_minutes'] + row['total_night_calls'] + row['total_night_charge'] + row['total_intl_minutes'] + row['total_intl_calls'] + row['total_intl_charge'], axis = 1)
```

```python
df_churn['Total Transaction'].describe()
```

```python
#2. State dari yang paling berdasarkan kelas

df_churn.groupby('state').sum()[['class']]
```

```python
## 3. Feature apa yang paling berkorelasi dengan column 'Class'

df_churn[df_churn.columns[1:]].corr()['class'][:].sort_values()

#Karena yang 0 yang tidak churn, yang menyebabkan customer stay adalah yang paling berkorelasi negatif
```

```python
#4. Statistik berdasarkan class (0 dan 1)

churn_yes = df_churn[df_churn['class']==1]
churn_no = df_churn[df_churn['class']==0]
```

```python
#ambil numeric column yg jadi measure saja
df_churn.select_dtypes(exclude='object').columns
```
