```python
#1. Customer yang tidak melakukan service sama sekali


df_churn['Total Transaction'] = df_churn.apply(lambda row: row['total_day_minutes'] + row['total_day_calls'] + row['total_day_charge'] + row['total_eve_minutes'] + row['total_eve_calls'] + row['total_eve_charge'] + row['total_night_minutes'] + row['total_night_calls'] + row['total_night_charge'] + row['total_intl_minutes'] + row['total_intl_calls'] + row['total_intl_charge'], axis = 1)
```
