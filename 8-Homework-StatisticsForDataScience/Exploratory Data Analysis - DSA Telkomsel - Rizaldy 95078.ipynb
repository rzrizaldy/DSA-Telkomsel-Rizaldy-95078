## B. Data Cleaning

```python
df[df['Market Category'].isna()]
```

```python
df = df.dropna(subset=['Market Category'])
```

```python
(1-(df.isnull().sum()/11914))*100
```

```python
#split caterogical into multiple columhn

# new data frame with split value columns 
new = df["Market Category"].str.split(",", n = 2, expand = True) 
```

```python
pd.options.mode.chained_assignment = None

df["Market Category 1"]=new[0] 
  
df["Market Category 2"]=new[1]

df["Market Category 3"]=new[2]
```

```python
df
```

```python
df[df['Engine Cylinders'].isna()]
#all engine cyclinders that NA are electric vehicle and rotary engine, replace value with 0 since electric vehicle have different engine

df[['Engine Cylinders']] = df[['Engine Cylinders']].fillna(value=0)
```

```python
df[df['Number of Doors'].isna()]

#number of doors assumed 4 for tesla since it's sedan, and 2 for ferrari FF since it's coupe

df[['Number of Doors']] = df[['Number of Doors']].fillna(value=4)
```

```python
df[df['Engine HP'].isna()]
```

```python
print(abs(df.corr()['Engine HP']).sort_values(ascending=False))
```

```python
#cari median dari kolom Engine HP dari top aspek yang berkorelasi dengan engine hp
#kolom yang diambil: cylinders, MSRP, mpg (city dan highway)


missing_enginehp = df[df['Engine HP'].isnull()]
mean_enginehp = df.groupby(['Engine Cylinders','MSRP','city mpg','highway MPG'])['Engine HP'].mean()

```

```python
df['Engine HP'] = df['Engine HP'].fillna(df.groupby(['Engine Cylinders','MSRP','city mpg','highway MPG'])['Engine HP'].transform('mean'))
```

```python
car_col = df[['city mpg','Engine HP','MSRP']]
```

```python
car_col.head()
```

```python
from sklearn import preprocessing

# Create a minimum and maximum processor object
min_max_scaler = preprocessing.MinMaxScaler()

# Create an object to transform the data to fit minmax processor
car_col_scaled = min_max_scaler.fit_transform(car_col)

```

```python
# Run the normalizer on the dataframe
df_scoring = pd.DataFrame(car_col_scaled)
```

```python
df_scoring.columns = ['mpg_z','enginehp_z','msrp_z']
```

```python
df["mpg_z"]=df_scoring["mpg_z"]
  
df["enginehp_z"]=df_scoring["enginehp_z"]

df["msrp_z"]=df_scoring["msrp_z"]
```

```python
df["score"]=(df['MSRP_zscore']+1)/(((df['Engine HP_zscore']+1)+(df['city mpg_zscore']+1)/2))
```

```python
df["score"]=df['msrp_z']/((df['enginehp_z']+df['mpg_z'])/2)
```

```python
df.describe()
```

```python
df_scoresort=df.groupby(by='Make')['score'].mean().sort_values(ascending=False)
result = pd.DataFrame(df_scoresort)
result
```
