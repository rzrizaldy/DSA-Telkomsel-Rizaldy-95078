# Data and Library Load

```python
import pandas as pd
import numpy as np
```

```python
gplay = pd.read_csv("googleplaystore.csv")
gplay.head()
```

```python
#column&row
print("column:",len(gplay.columns))
print("row:",len(gplay))
```

## Data Cleansing

```python
gplay.columns = gplay.columns.str.replace(' ', '_')
```

```python
gplay.Size=gplay.Size.str.replace('k','e+3')
gplay.Size=gplay.Size.str.replace('M','e+6')
gplay.Size.head()
gplay.Size=gplay.Size.replace('Varies with device',np.nan)
gplay.Size=gplay.Size.replace('1,000+',1000)
gplay.Size=pd.to_numeric(gplay.Size)
```

```python
gplay.Installs=gplay.Installs.apply(lambda x: x.strip('+'))
gplay.Installs=gplay.Installs.apply(lambda x: x.replace(',',''))
gplay.Installs=gplay.Installs.replace('Free',np.nan)
gplay.Installs.value_counts()
gplay.Installs=pd.to_numeric(gplay.Installs)
```

```python
gplay[~gplay.Reviews.str.isnumeric()]
```

```python
gplay=gplay.drop(gplay.index[10472])
```

```python
gplay.Reviews=gplay.Reviews.replace(gplay.Reviews[~gplay.Reviews.str.isnumeric()],np.nan)
```

```python
gplay.Price=gplay.Price.apply(lambda x: x.strip('$'))
```

```python
sep = ';'
rest = gplay.Genres.apply(lambda x: x.split(sep)[0])
gplay['Pri_Genres']=rest
gplay.Pri_Genres.head()
```

```python
rest = gplay.Genres.apply(lambda x: x.split(sep)[-1])
rest.unique()
gplay['Sec_Genres']=rest
gplay.Sec_Genres.head()
```

```python
twowaytable = pd.crosstab(index=gplay["Pri_Genres"],columns=gplay["Sec_Genres"])
twowaytable.head()
```
