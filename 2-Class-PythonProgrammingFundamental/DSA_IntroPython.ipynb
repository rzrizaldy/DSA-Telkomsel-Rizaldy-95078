```python
import pandas as pd
```

```python
code = ['US', 'AUS', 'JPN', 'IN', 'RUS', 'MOR', 'EG']
names = ['United States', 'Australia', 'Japan', 'India',
'Russia','Morocco', 'Egypt']
dr = [True, False, False, False, True, True, True]
cpc = [809, 731, 588, 18, 200, 70, 45]
# Create dictionary my_dict with three key:value pairs: my_dict
my_dict = {'ctrcode':code, 'country':names, 'drives_right':dr,
'cars_per_cap':cpc}
cars = pd.DataFrame(my_dict)
print(type(cars))
print(cars)
```

```python
# Dimension
cars.shape
print(cars.columns)
print(len(cars.columns))
print(len(cars))
# Brief View
cars.head(7)
cars.tail(3)
# Data type
cars.dtypes
cars.info()
```

```python
# Summary Statistics
cars.describe()
# Set DF Column Index
cars_id = cars.set_index("ctrcode")
print(cars_id.shape)
cars_id
```

```python
print(cars_id[cars_id.drives_right == True])
print(cars_id[(cars_id.cars_per_cap > 100) & (cars_id.country == "Australia") ])
```

```python
cars_id.sort_values(by=['cars_per_cap’])
cars_id.sort_values(by=['cars_per_cap'], ascending=False)
```
