## Manipulate Data with Pandas

### Series

Definition:<br>
- Series is 1 dimensional in nature such as an array. Series is a mutable data object. It means it can be updated, new items can be added and existing items can be deleted from the collection.
- It can hold data of any type.
- It can be instantiated with an array or a dictionary. The keys of the dictionary are used to represent indexes.
- The constructor of Series can be called by passing in data, index, data type and a Boolean flag indicating if we want to copy the data. If we don’t specify the index names then the indexes are labeled as integers starting from 0. <br>


$\textbf{Think of Series as Vertical Columns that can hold multiple rows.}$






#### Create

```python
#Creating an empty Series
s = pd.Series()
print(s)
```

```python
#From Python List
# Creating a Series from python list with default index values
list1 = [12,24,32]
s4 = pd.Series(list1)
print(s4)
```

```python
#with custom index values
s5 = pd.Series(list1, index = ['Peter', 'John', 'Mathew'])
print(s5)
```

```python
# From NumPy Array
# Creating a Series from a numpy array with default index values
import numpy as np
arr = np.array([12,24,32])
s = pd.Series(arr)
print(s)
```

```python
# From Python Dictionary
# Creating a Series from python dictionary
dict1 = {'Peter':12, 'John':24, 'Mathew':32}
s3 = pd.Series(dict1)
print(s3)
```

#### Retrieve

```python
#Retrieving data using position
#retrieve the first element
print(s4[0])

#retrieve the first three element
print(s4[:3])
```

```python
#Retrieving data using label
#retrieve a single element
print(s5['Peter'])
#retrieve multiple elements
print(s5[['Peter','John','Mathew']])
```

```python
# using loc and iloc
print(s5.loc['Peter'])

print(s5.iloc[0])
```

```python
#If a label is not contained, an exception is raised.
print(s5['Alex'])
```

### DataFrame

Definition:<br>
- Possibly the most used data structure in a data science project.
- It is a table with rows and columns — like a SQL Table Or Excel Spreadsheet Sheet.
- The table can store in memory data objects in different format.
- Offers high performing time series data analysis and engineering
- A data frame is a container of one or more Series.
- DataFrame is mutable. <br>

#### Create

```python
d = {'ColumnA' : pd.Series([111, 222, 333]),
   'ColumnB' : pd.Series([444, 555, 666]),
    'ColumnC' : pd.Series([777, 888, 999])}

df = pd.DataFrame(d)
df
```

#### Selection

```python
#Column Selection
#We can use the column name to get the data:
data = df['ColumnA']
data
```

```python
#multiple columns
data = df.iloc[:, :2]
print(data)

#or using label
data = df[['ColumnA', 'ColumnB']]
print(data)
```

```python
# Row Selection
data = df[:1]
print(data)

# or using iloc
data = df.iloc[0]
print(data)
```

```python
#multiple rows
data = df.iloc[:2, :]
print(data)
```

#### Addition

```python
# addition
# column addition
df['ColumnD'] = [123, 234, 567]
df
```

```python
# row addition
add = pd.DataFrame([[12,0, 3, 4]],columns = ['ColumnA','ColumnB', 'ColumnC', 'ColumnD'])
df.append(add)
```

#### Deletion

```python
# Deletion
# delete column
df.drop('ColumnA', axis = 1)
```

```python
# delete row
df.drop(0)
```

#### Rename

```python
df.rename(columns={'ColumnA': 'A'})
```

```python
df.rename(index={0: 'index0'})
```

#### Looping

```python
# loop over columns
for column in df.columns:
    print(column)
```

```python
# loop over items
for column,items in df.iteritems():
    print(column, items)
```

```python
# loop over rows:
for index_of_row, row in df.iterrows():
    print(index_of_row, row)
```

#### Sorting

```python
# sort by index
df.sort_index(ascending=False)
```

```python
# sort by columns
df.sort_values('ColumnD', ascending = True)
```

## Simple EDA

```python
import pandas as pd

def describe_data(data):
    print('length data: {}'.format(len(data)))
    print('shape of data: {}'.format(data.shape))
    print('info on data: ')
    print(data.info())
    print('describe data')
    print(data.describe())
    return
```

### What's the data looks like?

```python
data = pd.read_csv('csv/telco_customer.csv')
describe_data(data)
data.head()
```

### How many customer who churn?

```python
data.Churn.value_counts()
```

```python
churn = data.Churn.value_counts().reset_index(name='customers')
churn = churn.rename(columns={'index':'IsChurn'})
churn.plot.bar(x='IsChurn', y='customers')
```

### Male or Female, who mostly churn?

```python
churn_gender = data.groupby(['gender', 'Churn'])['customerID']\
.nunique().reset_index(name='unique_cust')
churn_gender
```

```python
churn_partner = data.groupby(['Partner', 'Churn'])['customerID'].nunique().reset_index(name='unique_cust')
churn_partner
```

```python
partner = data.groupby('Partner')['customerID'].nunique().reset_index(name = 'total_user')
churns = churn_partner.merge(partner, how = 'inner', on = 'Partner')
churns['percentage'] = churns['unique_cust']*100.0/churns['total_user']
churns
```

### How's the correlation between numbers in the data?

```python
import numpy as np
data['TotalCharges'] = np.where(data['TotalCharges'] == ' ',\
                                0, data['TotalCharges']) 
data['TotalCharges'] = data['TotalCharges'].astype(float)
data.corr()
```
