```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.dates as dates
import seaborn as sns
from IPython.core.pylabtools import figsize

%matplotlib inline
```

# Seaborn

### Scatter Plot

```python
iris = sns.load_dataset("iris")
iris.head(10)
```

```python
sns.scatterplot(x= 'sepal_length',
                y = 'sepal_width', 
                hue = 'species',
                data = iris).set_title("sepal_length vs sepal_width")
plt.show()
```

```python
sns.relplot(x= 'sepal_length',
            y = 'sepal_width', 
            kind = "scatter",
            data = iris)

plt.title('test title')
```

```python
sns.scatterplot(x= 'sepal_length', y = 'sepal_width',
                hue='species', style='species', data = iris).set_title("Scatter Plot Example")
plt.show()
```

```python
tips = sns.load_dataset('tips')
tips.head(10)
```

```python
sns.relplot(x="total_bill", y="tip", hue="smoker",
            kind = 'scatter',row="time", col = 'sex', data=tips)
plt.title("test_title")
```

### Line plot

```python
fmri = sns.load_dataset("fmri")
fmri.head(3)
```

```python
sns.lineplot(x="timepoint", y="signal",ci = 95, data=fmri).set_title("lineplot Plot Example")
plt.show()
```

```python
sns.lineplot(x="timepoint", y="signal", 
             hue = 'event', 
             data=fmri).set_title("lineplot Plot Example")
plt.show()
```

```python
sns.lineplot(x="timepoint", y="signal", 
             hue = 'event', data=fmri,
             ci = None).set_title("lineplot Plot Example")
plt.show()
```

```python
sns.lineplot(x="timepoint", y="signal",
             hue="event", style="event",
             markers=True, dashes=True, 
             data=fmri)
plt.show()
```

```python
sns.lineplot(x="timepoint", y="signal",
             hue="event",
             style="event",
             markers=True, dashes=False, 
             ci = None,
             data=fmri)
plt.show()
```

### Histogram Plot

```python
x = np.random.normal(size=30)
sns.distplot(x).set_title("Distribution Plot Example")
```

```python
sns.distplot(x, bins = 10).set_title("Distribution Plot Example")
```

```python
sns.distplot(x, kde = False, bins = 10, rug = True).set_title("Distribution Plot Example")
```

### KDE Plot

```python
sns.distplot(x, hist=False, rug=False).set_title("KDE Plot Example")
```

```python
sns.kdeplot(x, shade=True).set_title("KDE Plot Example")
```

### Multiplot Grid

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
g = sns.FacetGrid(tips, col="time")
```

```python
g = sns.FacetGrid(tips, col="time")
g.map(plt.hist, "tip");
```

```python
g = sns.FacetGrid(tips, col="time")
g.map(sns.distplot, "tip");
```

```python
g = sns.FacetGrid(tips, row="day", col = 'time')
```

```python
g = sns.FacetGrid(tips, row="day", col = 'time')
g.map(sns.distplot, "tip")
```

```python
g = sns.FacetGrid(tips, col="sex", hue="smoker")
g.map(plt.scatter, "total_bill", "tip", alpha=.7)
g.add_legend();
```
